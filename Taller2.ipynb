# Comparación de Series y Transformadas de Fourier

---

## 1. Serie de Fourier (Señales periódicas en tiempo continuo)

### Forma Trigonométrica

$$
x(t) = a_0 + \sum_{n=1}^{\infty} \left[ a_n \cos(n\omega_0 t) + b_n \sin(n\omega_0 t) \right]
$$

$$
\omega_0 = \frac{2\pi}{T}
$$

### Forma Exponencial

$$
x(t) = \sum_{n=-\infty}^{\infty} C_n e^{j n \omega_0 t}
$$

$$
C_n = \frac{1}{T} \int_{t_0}^{t_0 + T} x(t) e^{-j n \omega_0 t} dt
$$

### Forma Compacta

Es equivalente a la forma exponencial, pero con notación simplificada para facilitar cálculos.

**Características:**
- Señal: continua en el tiempo
- Frecuencia: espectro discreto
- Requiere que la señal sea periódica

---

## 2. Transformada de Fourier (TF)

$$
X(f) = \int_{-\infty}^{\infty} x(t) e^{-j 2\pi f t} dt
$$

$$
x(t) = \int_{-\infty}^{\infty} X(f) e^{j 2\pi f t} df
$$

**Características:**
- Señal: continua en el tiempo
- Frecuencia: espectro continuo
- Aplica para señales aperiódicas

---

## 3. DTFT – Transformada de Fourier en Tiempo Discreto

$$
X(e^{j\omega}) = \sum_{n=-\infty}^{\infty} x[n] e^{-j\omega n}
$$

$$
x[n] = \frac{1}{2\pi} \int_{-\pi}^{\pi} X(e^{j\omega}) e^{j\omega n} d\omega
$$

**Características:**
- Señal: discreta (infinita o aperiódica)
- Frecuencia: espectro continuo y periódico en \( \omega \), con período \( 2\pi \)

---

## 4. DFT – Transformada Discreta de Fourier

$$
X[k] = \sum_{n=0}^{N-1} x[n] e^{-j \frac{2\pi}{N}kn}, \quad k = 0, 1, \dots, N-1
$$

$$
x[n] = \frac{1}{N} \sum_{k=0}^{N-1} X[k] e^{j \frac{2\pi}{N}kn}, \quad n = 0, 1, \dots, N-1
$$

**Características:**
- Señal: discreta y finita (o periódica con período \( N \))
- Frecuencia: espectro discreto y periódico

---

## 5. Fast Fourier Transform (FFT)

### ¿Qué es?

Es un algoritmo eficiente para calcular la DFT. El más conocido es el algoritmo de Cooley–Tukey.

### Ventaja computacional

$$
\text{Costo DFT: } O(N^2)
$$

$$
\text{Costo FFT: } O(N \log_2 N)
$$

### Funcionamiento

Se divide la señal original \( x[n] \) de longitud \( N \) (siendo \( N = 2^m \)) en:

- Muestras pares: \( x[2n] \)
- Muestras impares: \( x[2n+1] \)

Se calcula la DFT recursivamente sobre cada mitad y se combinan con factores de giro:

$$
X[k] = E[k] + W_N^k \cdot O[k]
$$

$$
X[k + N/2] = E[k] - W_N^k \cdot O[k]
$$

$$
W_N^k = e^{-j \frac{2\pi k}{N}}
$$

### Aplicaciones

- Procesamiento de audio, voz, radar
- Análisis espectral en tiempo real
- Compresión de señales (MP3, JPEG)

## Transformada de Fourier — Densidad Espectral
La definición de la Transformada de Fourier para una función
$x(t)$ es:

$$
X(f) = \int_{-\infty}^{+\infty} x(t) \, e^{-j 2 \pi f t} \, dt
$$

a) $x(t) = e^{-a|t|}, \ a \in \mathbb{R}^{+}$

Se evalúa la integral separando en dos tramos:
$$
X(f) = \int_{-\infty}^{0} e^{(a - j 2 \pi f) t} \, dt + \int_{0}^{+\infty} e^{-(a + j 2 \pi f) t} \, dt
$$

Resolviendo e integrando:
$$
X(f) = \frac{1}{a - j 2 \pi f} + \frac{1}{a + j 2 \pi f}
$$
Sumando ambas fracciones:
$$
X(f) = \frac{2a}{a^2 + (2 \pi f)^2}
$$

b) $x(t) = \cos(w_c t), \ w_c \in \mathbb{R}$
